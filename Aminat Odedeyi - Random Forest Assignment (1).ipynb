Question: Use the iris and diabetes data set on sklearn build a random forest tree model

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
iris_data = load_iris()
```

```python
#1. Random forest tree model for iris dataset
iris_data
```

```python
iris_data.data
```

```python
#checking directory of iris dataset
dir(iris_data)
```

```python
#converting iris dataset to a dataframe
iris_df = pd.DataFrame(iris_data.data)
iris_df.head()
```

```python
iris_df["target"] = iris_data.target
iris_df
```

```python
iris_df[0:15]
```

Training, modeling and predicition

```python
x = iris_df.drop("target", axis = "columns")
y = iris_df.target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
model = RandomForestClassifier (n_estimators = 20)
model.fit(x_train, y_train)
```

```python
y_test
```

```python
x_test
```

```python
model.score(x_test, y_test)
```

```python
y_predicted_value = model.predict(x_test)
y_predicted_value
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
iris_conf_max = confusion_matrix(y_test, y_predicted_value)
iris_conf_max
```

```python
plt.figure(figsize = (9,6))
sns.heatmap(iris_conf_max, annot = True)
plt.xlabel("Predicted Value")
plt.ylabel("True Value")
```

```python
#2.Random forest tree model for diabetes dataset
from sklearn.datasets import load_diabetes
diabetes_data = load_diabetes()
diabetes_data
```

```python
#checking directory of diabetes dataset
dir(diabetes_data)
```

```python
#converting diabetes dataset to a dataframe
diab_df = pd.DataFrame(diabetes_data.data)
diab_df.head()
```

```python
diab_df.info()
```

```python
diab_df["target"] = diabetes_data.target
diab_df
```

Training, modeling and predicition

```python
x = diab_df.drop("target", axis = "columns")
y = diab_df.target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
model = RandomForestClassifier (n_estimators = 20)
model.fit(x_train, y_train)
```

```python
x_test
```

```python
y_test
```

```python
model.score(x_test, y_test)
```

```python
y_predicted_value = model.predict(x_test)
y_predicted_value 
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
diab_conf_max = confusion_matrix(y_test, y_predicted_value)
diab_conf_max
```

```python
plt.figure(figsize = (9,5))
sns.heatmap(diab_conf_max, annot = True)
plt.xlabel("Predicted Value")
plt.ylabel("True value")
```
